## ANSWERING THE QUESTION

5.0 AS A BUSINESS MANAGER, TRY TO FIND OUT THE WEAK AREAS WHERE YOU CAN WORK TO MAKE MORE PROFIT. WHAT ALL BUSINESS PROBLEMS CAN YOU DERIVE BY EXPLORING THE DATA?

```python
# Check profit by state
stateprofit=retail.groupby(["State"])["Profit"].sum().nlargest(15)
stateprofit
```

```python
plt.figure(figsize=(10,5))
stateprofit.plot.bar()
plt.title("Profit by State", fontsize= "12")
plt.xlabel("State")
plt.ylabel("Profit")
plt.show()
```

**The chart above illustrates the 15 highest profit-making states. California makes the highest profit of over 70000 but less than 80000, followed by New York. It also shows there is a large drop in profit. The superstore could stop selling to states that make very little or no profit and focus on the profit-making states.**

```python
# Check for Relationship Between quantity and discount, as against different segments
profit=retail.groupby("Segment")[["Quantity","Discount"]].sum().sort_values(by="Quantity",ascending=False)
profit
```

```python
profit=retail.groupby("Segment")[["Quantity","Discount"]].sum().sort_values(by="Quantity",ascending=False)
profit[:].plot.bar(figsize=(10,8), fontsize="12")
plt.title("Quantity and Discount by Segment")
plt.show()
```

**The chart illustrates the positive linear relationship between discounts and quantities. For the Home Office Segment with the least amount of quantities and discounts, the superstore could intoduce an incentive discount to customers who purchase in large quantities from this segment.**

```python
# Check for Sales and Profit By Region
sales_profit_by_region=retail.groupby("Region")[["Sales","Profit"]].sum().sort_values(by="Sales", ascending=False)
sales_profit_by_region
```

```python
sales_profit_by_region=retail.groupby("Region")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False)
sales_profit_by_region[:].plot.bar(figsize=(10,8), fontsize="12")
plt.title("Sales and Profit by Region")
plt.ylabel("Amount")
plt.show()
```

**The chart above illustrates sales and profit rates across the different regions. High amount of sales equals hight amount of profit. The West Region has high sales and profit. The superstore could focus their sales on the top three profit-making regions**

```python
# Check sales and profit of sub-category
sales_profit_by_subcategory=retail.groupby("Sub-Category")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False)
sales_profit_by_subcategory
```

```python
sales_profit_by_subcategory=retail.groupby("Sub-Category")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False)
sales_profit_by_subcategory[:].plot.bar(figsize=(10,8), fontsize="12")
plt.title("Sales and Profit by Sub-Category")
plt.ylabel("Amount")
plt.show()
     
```

**The above graph illustrates the profit and sales across different sub-category. Phones make the highest sales and profit. While tables, bookcases and suppliers make high sales(compared to labels, fastners and art), but all make losses. The superstore could remove the sales of bookcases, tables and suppliers as their sales do not produce profits.** 

```python
# Check profit by ship mode
profit_by_shipmode=retail.groupby(["Ship Mode"])["Profit"].sum().nlargest()
profit_by_shipmode
```

**The table above shows the different ship modes as a means of delivery. The highest profit comes from the standard class and the least come from Same Day.**
