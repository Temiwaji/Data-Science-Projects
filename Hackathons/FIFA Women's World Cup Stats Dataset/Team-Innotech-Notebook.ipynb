### 3.4 FILL NULL/MISSING VALUES
Check distribution of the columns with missing with null to determine appropiate values to fill the null values

```python
# Define function for figsize
def figsize(width, height):
    plt.figure(figsize=[width, height])
    
#Define function for title, y_axis label and x_axis label.
def labels(title, x_axis, y_axis):
    plt.title(title, fontsize = 16)
    plt.ylabel(y_axis, fontsize = 14)
    plt.xlabel(x_axis, fontsize = 14)
```

```python
# Define plot function
def plot_hist(variable):
    figsize=(8, 5)
    plt.hist(df_1[variable], bins=20);
    labels(f'Distribution of {variable}', variable, 'Counts')
```

```python
plot_hist('possesion')
```

- The distribution of possesion is almost normal with a mean of 49.3 qnd median of 49.4. The mean will be used to fill the missing value.

```python
def splot(row, column, index):
    plt.subplot(row, column, index)

def plot_bar(variable):
    figsize=(8, 5)
    sns.countplot(data = df_1, x = variable, color = "#1f77b4");
    labels(f"Distribution of {variable}", variable, "Counts")
```

```python
plot_bar('yellow_cards')
```

- mean will be used to fill the null values in the 'yellow_cards' column.

```python
plot_bar('red_cards')
```

- mean will be used to fill the null values in the 'red_cards' column.

```python
import math
```

```python
# Find the average value of possession
inputs_possesion=round(df_1["possesion"].mean(), 1)
inputs_possesion
```

```python
# Replace null values with average value in possesion column
df_1["possesion"].fillna(inputs_possesion, inplace = True)
```

```python
# Define the function to replace nulls in the red_cards and yellow_cards column
def fill_null(column):
    replacement = math.floor(df_1[column].mean())
    return df_1[column].fillna(replacement, inplace = True)
```

```python
# Apply the fill_null function
fill_null("yellow_cards")
fill_null("red_cards")
```

```python
df_1.sample(6)
```

```python
# Change dtype of red and yellow cards columns to integer after filling null values
df_1.yellow_cards = df_1.yellow_cards.astype("int")
df_1.red_cards = df_1.red_cards.astype("int")
```

```python
# Create new variable "penalty_goals" from goals and non_penalty_goals
df_1['penalty_goals'] = df_1['goals'] - df_1['non_penalty_goals']
```

### 4.1 RELATIONSHIP ANALYSIS
#### Bivariate Exploration
- Which are the top 10 squads according to goals?
- Which are the top 10 squads according to consistency in participation?

```python
# Check top 10 squads by goals
figsize(8,5)
top10_by_goals = df_1.groupby("squad")["goals"].sum().sort_values(ascending = False)[:10]
top10_by_goals.plot(kind = "bar", color = "#1f77b4");
labels("Top 10 squads by goals", "Squad", "Goals")
```

```python
# Check top 10 squads by participation
figsize(8,5)
top10_participants = df_1.groupby("squad")["year"].count().sort_values(ascending = False)[:10]
top10_participants.plot(kind = 'bar', color = "#1f77b4")
labels("Top 10 squads by participation", "Squad", "Frequency of Participation");
```

__Observation__
- As expected, all the teams that participated consistently scored a high number of goals, except for Nigeria, who was replaced by England.

#### Multivariate Exploration
Relationship among 3 or mor variables

```python
# Use heatmap to show graphical representation of the correlation between variables
corr = df_1.corr()
ones = np.ones_like(corr, dtype="bool")
mask = np.triu(ones)
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="Blues", linewidths=.6, cbar_kws={"shrink":.9})
ax.xaxis.tick_bottom()
plt.title("Correlations heatmap for FIFA Women's World Cup dataset".upper(), fontdict={"fontsize": 18}, loc="left")
```

__Observation__
There is a strong correlation between some key variables such as 
- goals and minimum playing time
- penalty kicks attempted and penalty kicks made
- goals and assists

```python
# Use relationplot to show relationship between variables
sns.relplot(x= "year", y= "age",hue = "goals", data=df_1,
           palette = "Blues");
```

**Observation**
- The highest number of goals was 25 and it occured in 1991, 1995, 2003 and 2019 tournaments. As the years progressed, the average age of players who scored and contributed to the 25 goals mark also increased.
