    3.5 SPLIT COLUMNS

```python
# Split the DateTime column into Date and Time columns FOR MERGED DATA
smdata[['Date', 'Time']] = smdata['Date'].str.split(' ', 1, expand=True)

smdata.head()
```

```python
# Convert Time column to string type
smdata['Time'] = smdata['Time'].astype(str)

# Function to convert time to 24-hour format
def convert_to_24hr(time_str):
    try:
        # Check if pm is in the time string
        if 'pm' in time_str.lower():
            # Split the time at ":" and remove any non-numeric characters
            time_parts = [part.strip('pm').strip() for part in time_str.split(":")]
            
            # Convert hours to 24-hour format by adding 12
            hours = int(time_parts[0]) + 12
            if hours == 24:
                hours = 12  # Special case for 12 pm, which becomes 24 hr
            minutes = time_parts[1]

            # Return the time in 24-hour format
            return f"{hours:02d}:{minutes}"

        # Check if am is in the time string
        elif 'am' in time_str.lower():
            # Split the time at ":" and remove any non-numeric characters
            time_parts = [part.strip('am').strip() for part in time_str.split(":")]
            
            # Convert hours to 24-hour format, keeping midnight (12 am) as 00:00
            hours = int(time_parts[0])
            if hours == 12:
                hours = 0  # Special case for 12 am
            minutes = time_parts[1]

            # Return the time in 24-hour format
            return f"{hours:02d}:{minutes}"

        # If neither am nor pm is in the time string, assume it's already in 24-hour format
        return time_str

    except Exception:
        return None  # Handle any exceptions gracefully

# Apply the conversion function to the "Time" column
smdata['Time'] = smdata['Time'].apply(convert_to_24hr)

smdata.head(2)
```

#### 4.1 RELATIONSHIP ANALYSIS

- Which content type has the highest avetage imprssions?
- Which network has te highest engagements?

```python
# Check content by impressions

# Group the data by content and calculate the mean impressions
content_impressions = smdata.groupby("Content Type")["Impressions"].mean().sort_values(ascending=False)

# Plot the species by average body mass
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=content_impressions.index, y=content_impressions.values, color="#1f77b4")
plt.title("Content Type by Average Impressions")
plt.xlabel("Contents")
plt.ylabel("Average Impressions")
plt.xticks(rotation=45, ha="right")

# Add count values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=10, color="black", xytext=(0, 5),
                textcoords="offset points")
```

**Observation**

Polls as a content type have the highest average impressions.

```python
# Check network/platform with most engaging posts
# Group the smdata by sent by and calculate the most engagements on an average
network_engagement = smdata.groupby("Network")["Engagements"].mean().sort_values(ascending=False)

# Plot the personnel by average reactions
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=network_engagement.index, y=network_engagement.values, color="#1f77b4")
plt.title("Network with Most Engagements on An Average")
plt.xlabel("Network")
plt.ylabel("Total Engagements")
plt.xticks(rotation=45, ha="right")

# Add count values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=10, color="black", xytext=(0, 5),
                textcoords="offset points")

plt.show()
```
