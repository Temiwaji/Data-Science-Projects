```python
# Define the function to replace nulls in the sex column
def fill_null(df, column):
    replacement = df[column].mode().values[0]  # Calculate the mode and access the first value
    df[column].fillna(replacement, inplace=True)
```

```python
# Call the function to replace nulls in the "sex" column
fill_null(penguindata, "sex")
```

#### 4.1 RELATIONSHIP ANALYSIS

- Which species have the highest average body mass?
- Which sex has the highest average bill length?

```python
# Check species by average body mass

# Group the data by species and calculate the mean body mass
species_body_mass = penguindata.groupby("species")["body_mass_g"].mean().sort_values(ascending=False)

# Plot the species by average body mass
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=species_body_mass.index, y=species_body_mass.values, color="#1f77b4")
plt.title("Species by Average Body Mass")
plt.xlabel("Species")
plt.ylabel("Average Body Mass (g)")
plt.xticks(rotation=45, ha="right")

# Add count values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=10, color="black", xytext=(0, 5),
                textcoords="offset points")

plt.show()
```

**Observation**

The Gentoo specie has the highest average body mass.

```python
# Check make and female penguins by average bill length
# Group the data by sex and calculate the mean bill length
sex_bill_length = penguindata.groupby("sex")["bill_length_mm"].mean().sort_values(ascending=False)

# Plot the species by average body mass
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=sex_bill_length.index, y=sex_bill_length.values, color="#1f77b4")
plt.title("Sex by Average Bill Length")
plt.xlabel("Sex")
plt.ylabel("Average Bill Length (mm)")
plt.xticks(rotation=45, ha="right")

# Add count values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=10, color="black", xytext=(0, 5),
                textcoords="offset points")

plt.show()
```

Male penguins gave the longest average bill length.

```python
sex = penguindata.groupby("sex")[["flipper_length_mm", "body_mass_g"]].sum().sort_values(by="flipper_length_mm", ascending=False)

# Create the bar plot
ax = sex[:].plot.bar(figsize=(10, 8), fontsize="12")

plt.title("Flipper Length & Body Mass by Sex")

# Add values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=12, color="black", xytext=(0, 5),
                textcoords="offset points")

plt.show()
```

**Observation**

The male species of the penguins have a higher total sum of flipper lengths and body mass

```python
# Group the data by 'island' and 'species' and calculate the count
species_counts = penguindata.groupby(['island', 'species']).size().reset_index(name='count')

# Create a bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=species_counts, x='island', y='count', hue='species', palette=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Count of Different Species on Different Islands')
plt.xlabel('Island')
plt.ylabel('Count')

# Move the legend to the right
plt.legend(title='Species', loc='upper left', bbox_to_anchor=(1, 1))

# Add values on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="baseline", fontsize=12, color="black", xytext=(0, 5),
                textcoords="offset points")

plt.show()
```

**Observation**

The Gentoo species, which is the most abundant, is exclusive to Biscoe Island, while the Adelie species can be observed on all three islands. Additionally, the chinstrap species is solely present on Dream Island.

#### 5.0 SUMMARY

The penguin dataset comprises of 343 entries, with three recorded species: chinstrap, adelie, and gentoo. Notably, the gentoo species stands out with the highest population and is exclusively found on Biscoe Island. While no single island hosts all three species, both Biscoe and Dream Islands are home to at least two species, adding an interesting geographical dimension to penguin distribution.

Additionally, a slight gender imbalance among the penguins has been observed, with males outnumbering females by 4%. Beyond this, the male penguins exhibit several physical advantages, boasting longer bill lengths, higher body masses, and more extended flippers compared to their female counterparts. These findings shed light on both the diversity in penguin species across islands and the gender-related disparities in physical characteristics among these fascinating creatures.
