#### 4. Calculate state-wise :
#### a. Total number of popluation.
#### b. Total no. of population with different religions.

```python
censusdata.head(2)
```

```python
#a.
censusdata.groupby("State_name").Population.sum().sort_values()
```

```python
#b.

# Take a look at the columns in the dataframe (to check the religions listed)
censusdata.columns
```

```python
# sort by the hindus religion
censusdata.groupby("State_name")["Hindus", "Muslims", "Christians", "Sikhs", "Buddhists", "Jains"].sum().sort_values(by = "Hindus")
```
