#### 5. What is the maximum & minimum 'average_price' per year in england ?

```python
london_housingdata["year"] = london_housingdata.date.dt.year
```

```python
london_housingdata.head(2)
```

```python
# Filter for England only
df1 = london_housingdata[london_housingdata.area == "england"]
df1.head(5)
```

```python
# Group by year
# Find maximum average price for England only
df1.groupby("year").average_price.max()
```

```python
# Find minimum average price for England only
df1.groupby("year").average_price.min()
```

```python
# Find mean average price for England only
df1.groupby("year").average_price.mean()
```

#### 6. What is the Maximum & Minimum No. of Crimes recorded per area ?

```python
# Max. no. of crimes per area
london_housingdata.groupby("area").no_of_crimes.max()
```

```python
# Min. no. of crimes per area and sort in descending order
london_housingdata.groupby("area").no_of_crimes.min().sort_values(ascending = False)
```
