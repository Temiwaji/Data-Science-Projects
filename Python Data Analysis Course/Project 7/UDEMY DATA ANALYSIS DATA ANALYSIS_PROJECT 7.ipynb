#### 7. Show all courses of Graphic Design where the price is below 100 ?

```python
coursedata.head(2)
```

```python
# Remove all rows which contains Free and create a new dataframe 'coursedata1'
coursedata1 = coursedata[coursedata.price != "Free"]
```

```python
# Change the price column datatype to 'int'
coursedata1.price = coursedata1.price.astype(int)
```

```python
# Get courses where the price of graphic design is below 100
coursedata1[(coursedata1.subject == "Graphic Design") & (coursedata1.price < 100)]
```

#### 9. What are courses that published in year 2015 ?

```python
# Check datatype of the columns
coursedata.dtypes
```

```python
# Convert dtype of 'published_timestamp' to 'datetime' format
coursedata["published_timestamp"] = pd.to_datetime(coursedata.published_timestamp)
```

```python
coursedata.dtypes
```

```python
coursedata.head(2)
```

```python
# Separate the year from the "published timestamp" column
coursedata["Year"] = coursedata["published_timestamp"].dt.year
```

```python
# Get courses published in year 2015
coursedata[coursedata.Year == 2015]
```
