### Question ( Groupby )
#### 3. Does gender affect who gets searched during a stop ?

```python
policedata.head(2)
```

```python
policedata.groupby("driver_gender").search_conducted.sum()
```

The above shows the number of times search was conducted on each gender

### Question ( mapping + data-type casting )

#### 4. What is the mean stop_duration ?

```python
policedata.head(2)
```

```python
# Check for unique values in the stop duration column
policedata.stop_duration.value_counts()
```

```python
policedata["stop_duration"] = policedata["stop_duration"].map({"0-15 Min" : 7.5, "16-30 Min" : 23, "30+ Min" : 45 })
```

```python
policedata
```

```python
# Get the mean of the stop duration coulmn
policedata["stop_duration"].mean()
```
